```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
file1 = pd.read_csv('files_for_activities/csv_files/regression_data1.csv')
data = pd.DataFrame(file1)
```

```python
def log_transfom_clean(x):
    if x != 0:
        if np.isfinite(x):
            return np.log(x)
        else:
            return 0
    else:
        return 0
data['IC3_log'] = list(map(log_transfom_clean, data['IC3']))
sns.displot(data['IC3_log'])
plt.show()
```

```python
iqr = np.percentile(data['IC3_log'],75) - np.percentile(data['IC3_log'],25)
upper_limit = np.percentile(data['IC3_log'],75) + 1.5*iqr
lower_limit = np.percentile(data['IC3_log'],25) - 1.5*iqr
```

```python
data1 = data[(data['IC3_log']>lower_limit) & (data['IC3_log']<upper_limit)]
```

```python
sns.distplot(data1['IC3_log'])
plt.show()
```

```python
iqr = np.percentile(data['IC3_log'],75) - np.percentile(data['IC3_log'],25)
upper_limit = np.percentile(data['IC3_log'],75) + 3*iqr
lower_limit = np.percentile(data['IC3_log'],25) - 3*iqr   
```

```python
data2 = data[(data['IC3_log']>lower_limit) & (data['IC3_log']<upper_limit)]
```

```python
sns.distplot(data2['IC3_log'])
plt.show()
```
