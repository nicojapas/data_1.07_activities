Repeat the same steps for column `IC5` for box-cox transformation.

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from scipy import stats
```

```python
file1 = pd.read_csv('files_for_activities/csv_files/regression_data1.csv')
data = pd.DataFrame(file1)
```

```python
data['IC5_'] = np.where(data['IC5']<=0,0,data['IC5'])
# now calculate mean of the new column , exluding zeros in the count 
mean = np.sum(data['IC5_'])/len(data[data['IC5_']>0])
data['IC5_'] = data['IC5_'].replace(0,mean)
```

```python
data = data.drop(['IC5'], axis=1)
```

```python
xt, lmbda = stats.boxcox(data['IC5_'])
```

```python
sns.distplot(xt)
plt.show()
```
