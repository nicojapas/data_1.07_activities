```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import Normalizer
```

```python
file1 = pd.read_csv('files_for_activities/csv_files/regression_data1.csv')
data = pd.DataFrame(file1)
```

```python
X_num = data.select_dtypes(include=np.number)
while X_num.shape[1]>3:
    X_num = X_num.drop(X_num.columns[-1],axis=1)
X_num.sort_values(by='IC1',ascending=False)
```

```python
# for i in range(X_num.shape[1]):
#     sns.distplot(X_num[X_num.columns[i]])

sns.set_style("ticks")
sns.displot(X_num, kde=True,stat="density",common_norm=False,height=5,aspect=1.5)
plt.show()
```

```python
transformer = Normalizer().fit(X_num)
x_normalized = transformer.transform(X_num)
x_normalized = pd.DataFrame(x_normalized)
```

```python
# for i in range(x_normalized.shape[1]):
#    sns.distplot(x_normalized[x_normalized.columns[i]])

sns.set_style("ticks")
sns.displot(data=x_normalized, kde=True,stat="density",common_norm=False, aspect=1.5)
plt.show()
```
